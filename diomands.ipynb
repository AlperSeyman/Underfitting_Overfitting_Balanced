```python
# Underfitting
# Overfitting
# Balanced

import seaborn as sns
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
df = sns.load_dataset("diamonds")
df.info()
```

```python
df.head()
```

```python
# All objects variable converted to int variable.
df = pd.get_dummies(df, columns=["cut","color","clarity"],drop_first=True,dtype="int")
```

```python
df.head()
```

```python
y = df["price"]
x = df.drop("price",axis=1)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=42, test_size=0.34)
```

```python
# Create Model
linear_reg = LinearRegression()
linear_model = linear_reg.fit(x_train, y_train)

# Model Test Score
linear_model.score(x_test, y_test)
```

```python
# Model Train Score
linear_model.score(x_train, y_train)
```
